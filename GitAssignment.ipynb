```python
# Task 1: Create a Git working directory with feature1.txt and feature2.txt in the master branch
mkdir git_assignment_2
cd git_assignment_2
git init
echo "This is feature 1" > feature1.txt
echo "This is feature 2" > feature2.txt
git add .
git commit -m "Initial commit with feature1.txt and feature2.txt"
```

```python
# Task 3: In develop branch create develop.txt, do not stage or commit it
git checkout develop
echo "This is the develop file" > develop.txt
```

```python
# Task 4: Stash this file and check out to feature1 branch
git stash --include-untracked
git checkout feature1
```

```python
# Task 5: Create new.txt file in feature1 branch, stage and commit this file
echo "This is the new file in feature1" > new.txt
git add new.txt
git commit -m "Add new.txt in feature1 branch"
```

```python
# Task 6: Checkout to develop, unstash this file and commit
git checkout develop
git stash pop
git add develop.txt
git commit -m "Add develop.txt in develop branch"
```

               Develop

               f2

3. Put develop.txt in develop branch, f1.txt and f2.txt in f1 and f2 respectively

4. Push all these branches to GitHub

5. On local delete f2 branch

6. Delete the same branch on GitHub as well3


```python
# Task 1: Create a Git working directory with the specified branches
mkdir git_assignment_3
cd git_assignment_3
git init
git branch Develop
git branch F1
git branch f2
```

```python
# Task 2: In the master branch, commit main.txt file
# Assuming you are currently on the master branch after git init
echo "This is the main file" > main.txt
git add main.txt
git commit -m "Add main.txt to master branch"
```

```python
# Task 3: Put develop.txt in develop branch, f1.txt and f2.txt in f1 and f2 respectively
git checkout Develop
echo "This is the develop file" > develop.txt
git add develop.txt
git commit -m "Add develop.txt to Develop branch"

git checkout F1
echo "This is the f1 file" > f1.txt
git add f1.txt
git commit -m "Add f1.txt to F1 branch"

git checkout f2
echo "This is the f2 file" > f2.txt
git add f2.txt
git commit -m "Add f2.txt to f2 branch"
```

```python
# Task 4: Push all these branches to GitHub
# Assuming you have a remote named 'origin' configured
# You will need to replace <your_remote_url> with your actual GitHub repository URL
# If you haven't added a remote, uncomment and run the following line first:
# git remote add origin <your_remote_url>

git push origin master
git push origin Develop
git push origin F1
git push origin f2
```

```python
# Task 5: On local delete f2 branch
git checkout master # or any other branch that is not f2
git branch -d f2
# Use -D instead of -d if the branch has unmerged changes you want to delete
# git branch -D f2
```

```python
# Task 6: Delete the same branch on GitHub as well
git push origin --delete f2
```
