Question :-  
 Compare and contrast mutable and immutable objects in Python with examples?
 ->
 In Python, objects can be classified as either mutable or immutable. The key difference lies in whether their state can be changed after they are created.

Mutable Objects:

Mutable objects can be modified after they are created. This means you can change their contents or elements without creating a new object.

Examples of mutable objects include:

Lists: You can add, remove, or change elements in a list.
Dictionaries: You can add, remove, or change key-value pairs in a dictionary
Immutable Objects:

Immutable objects cannot be modified after they are created. If you try to change an immutable object, a new object is created with the desired changes.

Examples of immutable objects include:

Strings: You cannot change individual characters in a string. Operations that appear to modify strings actually create new strings.

Tuples: You cannot add , remove , change elements in tuples

->

Python offers various types of operators to perform operations on variables and values.
1. Arithmetic Operators: Perform mathematical calculations.
+ (Addition): x + y
- (Subtraction): x - y
* (Multiplication): x * y

/ (Division): x / y (returns float)

% (Modulus): x % y (returns remainder)

Python

    a = 10

    b = 3

    print(a + b)  # Output: 13

    print(a // b) # Output: 3

2. Assignment Operators: Assign values to variables.
= (Assignment): x = 10
+= (Add and assign): x += 5 (equivalent to x = x + 5)
Other compound assignments like -=, *=, /=, %=, **=, //= exist.
Python

    x = 5

    x += 2

    print(x)  # Output: 7

3. Comparison (Relational) Operators: Compare two values and return a Boolean (True/False).
== (Equal to): x == y

!= (Not equal to): x != y

> (Greater than): x > y

< (Less than): x < y

>= (Greater than or equal to): x >= y

<= (Less than or equal to): x <= y

Python

    print(5 > 3)  # Output: True

    print(5 == 5) # Output: True
4. Logical Operators: Combine conditional statements.
and: Returns True if both operands are True.
or: Returns True if at least one operand is True.
not: Reverses the logical state of the operand.
Python

    print((5 > 3) and (10 < 20)) # Output: True

    print(not True)             # Output: False
5. Bitwise Operators: Perform operations on individual bits of integers.
& (Bitwise AND), | (Bitwise OR), ^ (Bitwise XOR), ~ (Bitwise NOT), << (Left shift), and >> (Right shift).
Python

    print(5 & 3)  # Binary: 0101 & 0011 = 0001 (Decimal: 1)
6. Identity Operators: Check if two variables refer to the same object in memory.
is: Returns True if both variables are the same object.
is not: Returns True if both variables are not the same object.
Python

    list1 = [1, 2, 3]

    list2 = list1

    list3 = [1, 2, 3]

    print(list1 is list2) # Output: True

    print(list1 is list3) # Output: False (different objects, same content)

in: Returns True if the value is found in the sequence.
not in: Returns True if the value is not found in the sequence.
Python

    my_list = [10, 20, 30]

    print(20 in my_list)     # Output: True
    
    print(5 not in my_list) # Output: True







Question :-  
 Explain the concept of type casting in Python with examples?

 ->
 Type casting, also known as type conversion, in Python refers to the process of converting a variable or value from one data type to another. This is often necessary when performing operations that require specific data types, or when data is received in a format different from what is needed for processing.
There are two main types of type conversion in Python:
Implicit Type Conversion (Coercion): This occurs automatically when Python converts one data type to another without explicit instruction from the programmer. This usually happens in expressions involving different numeric types, where Python promotes the "smaller" type to the "larger" type to avoid data loss.
Example:
Python

    integer_value = 10
    float_value = 5.5
    result = integer_value + float_value
    print(result)
    print(type(result))
In this example, the integer integer_value is implicitly converted to a float before the addition, resulting in a float result.

Explicit Type Conversion (Type Casting): This requires the programmer to explicitly convert a data type using built-in functions. These functions include int(), float(), str(), list(), tuple(), set(), etc. Explicit conversion is used when the desired type is not automatically inferred or when a specific conversion is required, even if it might involve potential data loss (e.g., converting a float to an integer truncates the decimal part).
Examples:
Python

    # Converting a string to an integer
    string_number = "123"
    integer_number = int(string_number)
    print(integer_number)
    print(type(integer_number))

    # Converting a float to an integer
    float_num = 7.89
    integer_from_float = int(float_num)
    print(integer_from_float) # Output: 7 (decimal part is truncated)
    print(type(integer_from_float))

    # Converting an integer to a string
    my_age = 30
    age_as_string = str(my_age)
    print(age_as_string)
    print(type(age_as_string))
Explicit type casting is crucial for ensuring data compatibility and enabling specific operations that are only valid for certain data types. For instance, mathematical operations require numerical types, while string manipulation functions operate on strings.

Questions :-  
 How do conditional statements work in Python? Illustrate with examples?
 ->

 Conditional statements in Python allow for the execution of specific code blocks only when certain conditions are met. This enables programs to make decisions and control their flow based on various inputs or circumstances.
1. if statement:
The if statement is the most basic conditional. It executes a block of code if a given condition evaluates to True.
Python

age = 20

if age >= 18:

    print("You are eligible to vote.")

In this example, since age (20) is greater than or equal to 18, the condition age >= 18 is True, and the message "You are eligible to vote." is printed.

2. if-else statement:
The if-else statement provides an alternative block of code to execute if the if condition is False.
Python

temperature = 25

if temperature > 30:

    print("It's a hot day!")

else:

    print("The weather is moderate.")

Here, temperature (25) is not greater than 30, so the else block is executed, printing "The weather is moderate."

3. if-elif-else statement:
The if-elif-else statement allows for checking multiple conditions sequentially. If the first if condition is False, it checks the elif (else if) conditions in order. If none of the if or elif conditions are met, the else block is executed.
Python

score = 75

if score >= 90:

    print("Grade: A")

elif score >= 80:

    print("Grade: B")

elif score >= 70:

    print("Grade: C")

else:
    print("Grade: F")
    
In this case, score (75) is not 90 or above, but it is 70 or above. Therefore, the elif score >= 70 condition is True, and "Grade: C" is printed.

Describe the different types of loops in Python and their use cases with examples?

->

Python offers two primary types of loops: for loops and while loops.
1. for Loop

Description: The for loop is used for iterating over a sequence (like a list, tuple, string, or range) or other iterable objects. It executes a block of code for each item in the sequence.

Use Cases:
Iterating through elements of a list or other collection.
Executing a block of code a specific number of times using range().
Processing characters in a string.

Example:

Python

    # Iterating through a list
    fruits = ["apple", "banana", "cherry"]
    for fruit in fruits:
        print(fruit)

    # Executing a specific number of times
    for i in range(5): # Iterates from 0 to 4
        print(f"Iteration {i}")
2. while Loop

Description: The while loop repeatedly executes a block of code as long as a given condition remains true. The condition is checked before each iteration.

Use Cases:

Repeating a task until a certain condition is met (e.g., user input validation).
Implementing game loops where the game continues until a specific event occurs.
Situations where the number of iterations is unknown beforehand.
Example:
Python

    count = 0
    while count < 3:
        print(f"Count is {count}")
        count += 1

    # User input validation
    password = ""
    while password != "secret":
        password = input("Enter the password: ")
        if password != "secret":
            print("Incorrect password. Try again.")
    print("Access granted!")
